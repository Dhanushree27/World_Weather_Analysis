```python
# Importing dependencies
import pandas as pd
import numpy as np
import requests
from citipy import citipy
import time
from datetime import datetime

# Importing API key
from config import weather_api_key
```

```python
# Generating random list of latitudes and longitudes
lats=np.random.uniform(-90,90,size=2000)
lngs=np.random.uniform(-180,180,size=2000)
lat_lngs=zip(lats,lngs)
coords=list(lat_lngs)
```

```python
# Identifying nearest city from lat,lng
cities=[]
for coord in coords:
    city=citipy.nearest_city(coord[0],coord[1]).city_name
    if city not in cities:
        cities.append(city)
```

```python
# Obtaining weather information for available cities
# Creating a list to store dictionaries
city_data=[]
# Setting up base URL
url="https://api.openweathermap.org/data/2.5/weather?units=imperial&appid="+weather_api_key+"&q="

# Looping through all cities
for i,city in enumerate(cities):
    
    # Sleeping to prevent time-out
    if (i>=50) & (i%50==0):
        time.sleep(60)
    try:        
        # Querying the data
        query_url=url+city.replace(" ","+")
        weather=requests.get(query_url).json()
        # Adding the data to a DF if city exists
        city_data.append({"City":city,
                        "Country":weather["sys"]["country"],
                        "Lat":weather["coord"]["lat"],
                        "Lng":weather["coord"]["lon"],
                        "Max Temp":weather["main"]["temp_max"],
                        "Humidity":weather["main"]["humidity"],
                        "Cloudiness":weather["clouds"]["all"],
                        "Wind Speed":weather["wind"]["speed"],
                        "Current Description":weather["weather"][0]["description"]
                        })
    # Skipping city if it does not exist
    except:
        print (f"City {city} not found. Skipping.....")
      
```

```python
# Creating a dataframe
weather_df=pd.DataFrame(city_data)
weather_df["City"]=weather_df["City"].str.capitalize()
weather_df.head()
```

```python
# Writing the dataframe to file
weather_df.to_csv("WeatherPy_Database.csv",index_label="City_ID")
```
